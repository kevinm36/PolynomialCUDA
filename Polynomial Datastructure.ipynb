```python
def split_coefficient_variable(string):
    n = len(string)
    for ind,c in enumerate(string):
        if c.isnumeric():
            if ind == n-1:
                return (string,u'0')
            else:
                continue
        else:
            return (string[0:ind],string[ind+1::])
```
